```python
import pandas, numpy
from pathlib import Path
from datetime import datetime
import json

CWD = Path("C:\\Users\\정진혁\\offline_main\\projectXpython")
SOURCE = CWD / 'source'
RESULT_STORAGE = CWD / 'result_storage'
MASSIVE_RESULT_STORAGE = Path('../MassiveResultStorage')

DATAS = Path("../datas")
ARTICLE_JSONS = DATAS / 'article_jsons'
TWITTER_JSONS = DATAS / 'twitter_jsons'

def print_log(message:str):
    print(f"{datetime.now()}\t{message}")

def import_file(file_path:Path):

    if file_path.suffix == ".xlsx":
        df = pandas.read_excel(file_path)
    elif file_path.suffix == ".csv":
        df = pandas.read_csv(file_path)
    else:
        print_log(f"skipped: {file_path} is not xlsx or csv file.")
    
    print_log(f"imported: {file_path}")

    return df

source_path = SOURCE / "catch_duplication"
result_path = RESULT_STORAGE / "catch_duplication"
reference_file = RESULT_STORAGE / "temp/catch_duplicate.json"
```

```python
"""
1. xlsx 파일, csv 파일을 각각 확장자별로 분리해 xl_df, csv_df에 저장한다.
2. reference_file은 catch_duplicate.json 파일이며, 이 파일에는 중복되는 Doc_ID의 리스트가 저장되어 있다.
3. xl_df에 csv_df의 데이터를 추가하되, reference_file에 저장된 Doc_ID에 대해서는 xl_df에 추가하지 않는다.
4. xl_df를 저장하되, xl_df의 행이 5만 개가 넘어가면, 5만 개씩 잘라서 result_path에 저장하고, 파일명을 "20231204_원천데이터(기사+한국관련기사)_중복제거_{i}.xlsx"로 저장한다.
"""
xl_df = pandas.DataFrame()
csv_df = pandas.DataFrame()
ref:list = json.load(reference_file.open(encoding="utf-8"))

xl_df_list = []
csv_df_list = []

for file_path in source_path.glob("**/*.*"):
    df = import_file(file_path)

    if file_path.suffix == ".xlsx": xl_df_list.append(df)
    elif file_path.suffix == ".csv": csv_df_list.append(df)
    else: continue

xl_df = pandas.concat(xl_df_list)
csv_df = pandas.concat(csv_df_list)

xl_df, csv_df
```

```python
# csv_df에서 ref에 있는 Doc_ID를 제거한다.
csv_df = csv_df[~csv_df["Doc_ID"].isin(ref)]
csv_df
```

```python

# xl_df에 csv_df를 추가한다.
xl_df = pandas.concat([xl_df, csv_df])

print_log("xl_df에 csv_df를 추가했습니다.")

xl_df
```

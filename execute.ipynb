```python
# JSONformat_handler 이용한 JSON 포맷 오류 해결 도구

from tool import JSONformat_handler

source_path = ARTICLE_JSONS
ban_pattern = r"statistic"
bp = re.compile(ban_pattern)

current_folder_name = ""

for json_path in natsorted(source_path.glob("**/*.json")):

    if bp.search(json_path.__str__()): continue

    if json_path.parent != current_folder_name:
        current_folder_name = json_path.parent
        common.print_log(f"Current folder: {current_folder_name}")

    data = common.read_json(json_path)
    
    data = JSONformat_handler.handle_format_exceptions(data)
    data = JSONformat_handler.handle_dtype_exceptions(data)
    data = JSONformat_handler.handle_tag_exceptions(data)
    data = common.put_entity(data)

    data = JSONformat_handler.arrange_json_format(data)

    common.write_json(data, json_path)
```
